# Caso de estudio 1: Flujo isentrópico en toberas

## Enunciado

### 1. Proceso de llenado de un tanque de aire comprimido

Tenemos un tanque de gas comprimido, conectado a una línea de presión con aire a presión $p_L$ y temperatura $T_L$, conectada al tanque mediante una válvula cuyo volumen es $V_T$, el cual contiene aire a una presión $p_T$ y temperatura $T_T$.

Asumiendo apertura instantánea de la válvula y sin pérdidas introducidas, se establece un flujo de aire de caudal variable en el tiempo, el cual disminuye a medida que el cilindro se llena.

Asumiendo que el proceso es adiabático y que las condiciones en la línea de alta presión son constantes en el tiempo, la presión, temperatura y velocidad en la entrada del conducto siguen las relaciones adiabáticas para una tobera convergente:

- $\frac{p_0}{p_g} = \left [ \frac{\gamma + 1}{2}\right ]^{\frac{\gamma}{\gamma - 1}}$
- $\frac{T_0}{T_g} = \left [ \frac{\gamma + 1}{2}\right ]$
- $u_g = a_g = \sqrt{\frac{2\gamma}{\gamma + 1}} RT_0$

Pueden plantearse algunas de las leyes de conservación inestacionarias para un volumen de control:

- Conservación de la masa
- Conservación de la cantidad de movimiento
- Conservación de la energía

en conjunto con la ecuación de estado para un gas ideal, y así resolver el valor del caudal másico, la presión, densidad y temperaturas instantáneas como funciones del tiempo en el tanque.

#### 1.1 Objetivos
- Determinar el tiempo que demora en llenarse el tanque hasta que $p_T = 0.95p_L$
- Graficar las curvas de caudal másico, presión, densidad y temperatura en el tanque en función del tiempo.

## Deducción de ecuaciones a utilizar

## Condición isentrópica, relaciones de potencia

Si consideramos el cambio de entropía utilizando las leyes primera y segunda de la termodinámica para una sustancia pura:

\begin{equation}
    T ds = dh - \frac{dp}{\rho}
\end{equation}

Si introducimos $dh = c_p dT$ para un gas perfecto, sustituímos $\rho T = p/R$ y despejando $ds$ e integrando entre dos estados 1 y 2:

\begin{equation}
    \int_1^2 ds = \int_1 ^2 c_p \frac{dT}{T} - R\int_1^2 \frac{dp}{p}
\end{equation}

Para $c_p$ constante:

\begin{equation}
    s_2 - s_1 = c_p ln \frac{T_2}{T_1} - Rln\frac{p_2}{p_1} = c_vln\frac{T_2}{T_1} - Rln\frac{\rho_2}{\rho_1}
\end{equation}

Si el flujo es isentrópico $s_1 = s_2$:

\begin{equation}
    \frac{p_2}{p_1} = \left(\frac{T_2}{T_1}\right)^{k/\left(k-1\right)} = \left(\frac{\rho_2}{\rho_1}\right)^k
\end{equation}

### Velocidad del sónido

De White podemos obtener el balance másico para una onda de presión de baja intensidad moviendose a través de un fluído estacionario a velocidad C, y mediante la ecuación de conservación de la masa se puede escribir:

\begin{equation}
    \rho A C = \left(\rho + \Delta \rho\right) A \left(C - \Delta V\right)
\end{equation}

Despejándo $\Delta V$:

\begin{equation}
    \Delta V = C \frac{\Delta \rho}{\rho + \Delta \rho}
\end{equation}

Al despreciar los esfuerzos viscosos debido a los bajos gradientes de velocidad, podemos escribir la 2da ley de Newton de la siguiente manera:

\begin{equation}
    \sum F_{rith} = \dot{m}\left(V_{out} - V_{in}\right)
\end{equation}

\begin{equation}
    pA - \left(p + \Delta p\right) A = \left(\rho A C\right) \left(C - \Delta V - C\right)
\end{equation}

operando las ecuaciones obtenemos:

\begin{equation}
    C^2 = \frac{\Delta p}{\Delta \rho} \left( 1 + \frac{\Delta \rho}{\rho}\right)
\end{equation}

y en el límite de $\Delta \rho \rightarrow 0$ (con la onda con una intensidad infinitesimal) tenemos la velocidad del sonido:

\begin{equation}
    a^2 = \cfrac{\partial p}{\partial \rho}
\end{equation}

La evaluación de la derivada debe ralizarse sobre un proceso inifitesimal adiabático o isentrópico, debido a los gradientes de temperatura inexistentes, salvo dentro de la misma onda. De esta manera, si consideramos un gas ideal, la derivada será:

\begin{equation}
 a =  \left(
        \left.\begin{matrix}
            \cfrac{\partial p}{\partial \rho}
        \end{matrix}
    \right|_s\right)^{1/2}
    = \left(
        \left.\begin{matrix}
           k \cfrac{\partial p}{\partial \rho}
        \end{matrix}
    \right|_T\right)^{1/2} 
\end{equation}

Finalmente de la ecuación de estado de los gases ideales:

\begin{equation}
    a =\left(\cfrac{kp}{\rho}\right)^{1/2} = \left(kRT\right)^{1/2}
\end{equation}

### Propiedades de flujo de gases en conductos

Analizando el flujo de un gas a lo largo de una pared aislada termicamente, podemos escribir la ecuación de la energía del flujo en la forma:

\begin{equation}
    h_1 + \frac{1}{2}V_1^2 + gz_1 = h_2 + \frac{1}{2}V_2^2 + gz_2 - q + w_v
\end{equation}

Si despreciamos las diferencias de elevación $\Delta z$ y asumimos flujo sin intercambio de trabajo ni calor debido a estar lejos de la pared, y por ende lejos de los límites donde la trasnmisión de calor y trabajo por efectos viscosos es significativa, podemos reducir la ecuación a:

\begin{equation}
    h_1 + \frac{1}{2}V_1^2 + gz_1 = h_2 + \frac{1}{2}V_2^2 + gz_2 = const
\end{equation}

El término $const$ es la entalpía de estancamiento, y es la entalpía del fluído cuando es llevado al reposo de manera adiabática. Es llamado también $h_0$:

\begin{equation}
    h + \frac{1}{2}V^2 = h_0 = const
\end{equation}

Para gases ideales:

\begin{equation}
    c_p T + \frac{1}{2}V^2 = c_p T_0
\begin{equation}

Dividiendo todo por $c_p T$ obtenemos:

\begin{equation}
    1 + \frac{V^2}{2c_p T} = \frac{T_0}{T}
\end{equation}

Pero de la ley de los gases ideales, $c_p T = \left [ kR / \left(k - 1\right)\right] T = a^2/\left(k - 1\right)$ y reemplazando tenemos:

\begin{equation}
    1 + \frac{\left(k-1\right)V^2}{2a^2} = \frac{T_0}{T}
\end{equation}

Apareciendo el número de Mach en la ecuación:

\begin{equation}
    \frac{T_0}{T} = 1 + \frac{k-1}{2}Ma^2
\end{equation}

Con $Ma = \cfrac{V}{a}$.

De la ecuación de la velocidad del sonido podemos deducir:

\begin{equation}
    \frac{a_0}{a} = \left(\frac{T_0}{T}\right)^{1/2} = \left[1 + \frac{1}{2}\left(k - 1\right)Ma^2\right]^{1/2}
\end{equation}

Por último, de las ecuaciones de flujo isentrópico podemos deducir:

\begin{equation}
    \frac{p_0}{p} = \left(\frac{T_0}{T}\right)^{k/\left(k-1\right)} = \left[1 + \frac{1}{2}\left(k-1\right)Ma^2\right]^{k/\left(k-1\right)}
\end{equation}

\begin{equation}
    \frac{\rho_0}{\rho} = \left(\frac{T_0}{T}\right)^{1/\left(k-1\right)} = \left[1 + \frac{1}{2}\left(k-1\right)Ma^2\right]^{1/\left(k-1\right)}
\end{equation}

Para $Ma = 1$ obtenemos las ecuaciones brindadas en el enunciado, que muestran los parámetros críticos o sónicos.

La velocidad crítica es por definición la velocidad del sonido, y puede escribirse como:

\begin{equation}
    \ast{V} = \ast{a} = \left(kR\ast{T}\right)^{1/2} = \left(\frac{2k}{k+1}RT_0\right)^{1/2}
\end{equation}

## Condición isentrópica

Si tomamos el diferencial de la ecuación de conservación de la energía simplificada con $h$ y $V$, tenemos:

\begin{equation}
    dh + VdV = 0
\end{equation}

y para un proceso isentrópico

\begin{equation}
    dh = \frac{dp}{\rho}
\end{equation}

Combinandolas:

\begin{equation}
    \frac{dp}{\rho} + VdV = 0
\end{equation}

Esta es la ecuación de Bernoulli para flujo sin fricción y estacionario con terminos de posición despreciables. Por lo que damos como válida la condición isentrópica.

### Conservación de la masa

Si asumimos el flujo en un ducto de sección variable, la velocidad estará descrita por un perfil $V\left(x,y\right)$. Sin embargo si el cambio del área es pequeño y el radio de curvatura de la pared es grande, entonces podemos decir que $V \approx V\left(x\right)$, reaccionando al cambio de área $A\left(x\right)$.

De esta manera, para flujo estacionario en una dimensión, la ecuación de continuidad será:

\begin{equation}
    \rho\left(x\right)V\left(x\right)A\left(x\right) = \dot{m} = const
\end{equation}

Calculando el diferencial de la función:

\begin{equation}
    \frac{d\rho}{\rho} + \frac{dV}{V} + \frac{dA}{A} = 0
\end{equation}

e incluyendo la definición de velocidad del sonido:

\begin{equation}
    dp = a^2 d\rho
\end{equation}

Si trabajamos con las tres ecuaciones llegamos a:

\begin{equation}
    \frac{dV}{V} = \frac{dA}{A}\frac{1}{Ma^2 - 1} = -\frac{dp}{\rho V^2}
\end{equation}

De aquí podemos deducir cómo existe una relación de cambio inversa entre propiedades para flujos supersónicos y subsónicos, debido al término $Ma^2 - 1$. Si el flujo es subsónico, y el área aumenta al avanzar el fluído, la presión aumenta y la velocidad disminuye, pero si el flujo es supersónico y el área aumenta, entonces la presión disminuye y la velocidad aumenta. Ahora si el flujo es subsónico y el área disminuye, entonces la presión disminuye y la velocidad aumenta, y lo contrario sucede si el flujo es supersónico. Tenemos los casos de difusor subsónico, tobera supersónica, tobera subsónica y difusor supersónico respectivamente.

Si igualamos el caudal en alguna región genérica del conducto al caudal en condiciones sónicas tenemos:

\begin{equation}
    \rhoVA = \ast{\rho}\ast{\ast{V}\ast{A}
\end{equation}

\begin{equation}
    \frac{A}{\ast{A}} = \frac{\ast{\rho}}{\rho}\frac{\ast{V}}{V}
\end{equation}

Esta última es una función cuyo máximo está en $Ma = 1$, por lo que se deduce que la condición de flujo máximo es aquella donde las condiciones sean sónicas.

Por otro lado, si usamos las ecuaciones obtenidas para la variación de la presión, y densidad en el conducto, obtenemos una ecuación de flujo másico:

\begin{equation}
    \dot{m} = \frac{Ap_0}{\sqrt{RT_0}}\sqrt{\frac{2k}{k-1}\left(\frac{p}{p_0}\right)^{2/k}\left[1-\left(\frac{p}{p_0}\right)^{\left(k-1\right)/k}\right]}
\end{equation}

Siendo $p$ y $A$ la presión y área en la sección de interés, y las propiedades con sub índice $0$ son las de estancamiento.
